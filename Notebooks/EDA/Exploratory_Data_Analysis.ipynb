### **What is the most popular recipe category (most repeated)?**<a id='Most_Popular_Recipe_Category'></a>
[Contents](#Contents)



```python
def most_popular_category(df):
    popular_category = df['category'].value_counts().head(10)
    plt.figure(figsize=(10, 6))
    ax = sns.barplot(x=popular_category.values, y=popular_category.index)
    plt.xlabel('Number of Recipes')
    plt.ylabel('Recipe Category')
    plt.title('Top 10 Most Popular Recipe Categories')
    
    # Add grid lines:
    plt.grid(axis='x', linestyle='--', alpha=0.7)
    
    # Add labels to the bars:
    for i in ax.containers:
        ax.bar_label(i, fmt='%d', label_type='edge', padding=2)
        
    # Adjust the layout to ensure labels fit within the frame
    plt.tight_layout()    
    plt.show()

# Call the function:
most_popular_category(df)
```

**Chart Highlights:**

The chart showcases the top 10 most popular recipe categories based on the number of recipes. "Appetizers-and-snacks" leads with the highest count, followed by "desserts" and "world-cuisine." The chart highlights the diversity and popularity of these categories, reflecting common culinary interests and preferences.


### **What is the average number of reviews/ratings?**<a id='Average_Number_of_Reviews_Ratings'></a>
[Contents](#Contents)

```python
def average_reviews_ratings(df):
    avg_reviews = df['review_count'].mean()
    avg_ratings = df['rating'].mean()
    
    # Data for plotting
    averages = {'Average Reviews': avg_reviews, 'Average Ratings': avg_ratings}
    
    # Plotting
    plt.figure(figsize=(10, 6))
    sns.barplot(x=list(averages.values()), y=list(averages.keys()), palette='pastel', hue=list(averages.keys()), dodge=False)
    plt.xlabel('Average Value')
    plt.ylabel('Metric')
    plt.title('Average Reviews and Ratings')
    plt.grid(axis='x', linestyle='--', linewidth=0.7)
    
    # Adding the labels on top of the bars
    for index, value in enumerate(averages.values()):
        plt.text(value - (value * 0.1), index, f'{value:.2f}', va='center', ha='center', color='black', weight='bold')
    
    plt.legend([],[], frameon=False)  # Remove the legend
    plt.show()

# Call the function
average_reviews_ratings(df)
```

**Chart Highlights:**
* Average Reviews: The average number of reviews for the recipes is `112.04`, indicating that recipes generally receive a significant amount of feedback from users.
* Average Ratings: The average rating across recipes is `4.07`, showing that the recipes are generally well-received by users.
* The contrast between the average number of reviews and ratings highlights the engagement level, with more users contributing to reviews than giving ratings.

### **What is the average number of reviews/rating count per category?**<a id='Average_Reviews_Rating_Count_per_Category'></a>
[Contents](#Contents)

```python
def average_reviews_ratings_per_category(df):
    avg_reviews_per_category = df.groupby('category')['review_count'].mean().sort_values(ascending=False).head(10)
    
    plt.figure(figsize=(10, 6))
    sns.barplot(x=avg_reviews_per_category.values, y=avg_reviews_per_category.index, hue=avg_reviews_per_category.index, dodge=False, palette='deep')
    
    plt.xlabel('Average Number of Reviews')
    plt.ylabel('Recipe Category')
    plt.title('Top 10 Categories by Average Reviews')
    plt.grid(axis='x', linestyle='--', linewidth=0.7)
    
    # Adding the labels on top of the bars
    for index, value in enumerate(avg_reviews_per_category.values):
        plt.text(value - (value * 0.1), index, f'{value:.2f}', va='center', ha='center', color='black', weight='bold')
    
    plt.legend([], [], frameon=False)  # Hide the legend
    plt.show()

# Call the function
average_reviews_ratings_per_category(df)

```

**Chart Highlights:**

The chart showcases the top 10 recipe categories based on the average number of reviews. Here are some highlights:

* Pasta and Noodles lead significantly with the highest average number of reviews, indicating strong popularity and engagement in this category.
* Meat and Poultry and Desserts also have high average reviews, reflecting their broad appeal and frequent preparation.
* Categories like Breakfast and Brunch, Fruits and Vegetables, and Drinks also feature prominently, suggesting diverse interest and frequent searches in these areas.
* World Cuisine and Main Dish categories show considerable average reviews, pointing to a wide range of interests in global flavors and main course meals.
* Soups, Stews, and Chili and Bread round out the top 10, highlighting substantial engagement in comfort foods and baking.

This chart helps identify which types of recipes are most engaging to users based on review activity, providing valuable insights for content focus and recommendation strategies.

### **What is the distribution of reviews/ratings?**<a id='What_is_the_distribution_of_reviews_ratings'></a>
[Contents](#Contents)

```python
def distribution_reviews_ratings(df):
    plt.figure(figsize=(10, 6))
    sns.histplot(df['review_count'], bins=50, kde=True, color='steelblue')
    plt.xlabel('Number of Reviews')
    plt.ylabel('Frequency')
    plt.title('Distribution of Reviews')
    plt.grid(axis='y', linestyle='--', linewidth=0.7)
    plt.show()

    plt.figure(figsize=(10, 6))
    sns.histplot(df['rating'], bins=50, kde=True, color='steelblue')
    plt.xlabel('Ratings')
    plt.ylabel('Frequency')
    plt.title('Distribution of Ratings')
    plt.grid(axis='y', linestyle='--', linewidth=0.7)
    plt.show()

# Call the function
distribution_reviews_ratings(df)
```

**Charts Highlights:**

* **Distribution of Reviews**:The chart displaying the distribution of reviews reveals a highly skewed distribution, with the majority of recipes receiving fewer reviews. There is a significant concentration of recipes with a minimal number of reviews, while a few recipes have received an exceptionally high number of reviews. This indicates that while some recipes are highly popular, many others are relatively less reviewed.

* **Distribution of Ratings**:
The distribution of ratings chart shows a more balanced distribution compared to reviews. Most recipes have ratings clustered around 4 to 5 stars, indicating generally favorable feedback from users. There are fewer recipes with lower ratings, suggesting that most recipes are well-received. The presence of a peak at 0 indicates a number of recipes with no ratings at all, which could be due to new or less popular recipes.

These insights can help identify popular and well-received recipes and also highlight areas where there might be opportunities to improve or promote recipes that have received fewer reviews and ratings.

### **Is there a correlation between recipe length and the number of reviews?**<a id='Is_there_a_correlation_between_recipe_length_and_the_number_of_reviews'></a>
[Contents](#Contents)

```python
def correlation_length_reviews(df):
    correlation = df[['recipe_length', 'review_count']].corr().iloc[0, 1]
    print(f'Correlation between Recipe Length and Number of Reviews: {correlation:.2f}')

    plt.figure(figsize=(10, 6))
    sns.scatterplot(x='recipe_length', y='review_count', data=df)
    
    # Adding the correlation line
    sns.regplot(x='recipe_length', y='review_count', data=df, scatter=False, color='red', line_kws={'linewidth': 1.5})

    plt.xlabel('Recipe Length (Number of Words)')
    plt.ylabel('Number of Reviews')
    plt.title('Recipe Length vs Number of Reviews')
    plt.grid(True, linestyle='--', linewidth=0.7)
    plt.show()

# Call the function
correlation_length_reviews(df)

```

**Chart Highlights:**
* Correlation Analysis: The correlation coefficient between recipe length and the number of reviews is -0.06, indicating a very weak negative correlation.
* Data Distribution: Most recipes cluster between 0 to 200 words in length.
* Review Counts: The majority of recipes receive fewer than 2000 reviews, with a few outliers receiving significantly higher numbers of reviews.
* Scatter Pattern: The scatter plot shows a dense concentration of data points in the lower range of recipe lengths and review counts, with fewer recipes having longer lengths or higher review counts.

Overall, the weak negative correlation suggests that longer recipes are slightly less likely to receive a higher number of reviews, but the relationship is not strong.

### **Is there a correlation between recipe length and average ratings?**<a id='Is_there_a_correlation_between_recipe_length_and_average_ratings'></a>
[Contents](#Contents)

```python
def correlation_length_ratings(df):
    correlation = df[['recipe_length', 'rating']].corr().iloc[0, 1]
    print(f'Correlation between Recipe Length and Average Ratings: {correlation:.2f}')

    plt.figure(figsize=(10, 6))
    sns.scatterplot(x='recipe_length', y='rating', data=df, s=50, alpha=0.6)
    
    # Adding the correlation line
    sns.regplot(x='recipe_length', y='rating', data=df, scatter=False, color='red', line_kws={'linewidth': 1.5})

    plt.xlabel('Recipe Length (Number of Words)')
    plt.ylabel('Average Ratings')
    plt.title(f'Recipe Length vs Average Ratings\nCorrelation: {correlation:.2f}')
    plt.grid(True, linestyle='--', linewidth=0.7)
    plt.show()

# Call the function
correlation_length_ratings(df)
```

**Chart Highlights:**
* Correlation: The correlation between recipe length and average ratings is -0.04, indicating a very weak negative relationship. This suggests that the number of words in a recipe's directions has almost no impact on its average rating.
* Distribution: Most recipes cluster around shorter lengths (less than 200 words) and higher ratings (between 3 and 5).
* Outliers: A few recipes are much longer (up to 700 words) and have varied ratings. These outliers do not significantly impact the overall weak correlation.
* Ratings Spread: Recipes with very short lengths (less than 50 words) show a wide range of ratings from 0 to 5, highlighting that even brief recipes can be highly rated.

Overall, the chart suggests that recipe length is not a significant factor in determining a recipe's rating.

### **What are the top 10 longest recipes (based on recipe length)?**<a id='What_are_the_top_10_longest_recipes_based_on_recipe_length'></a>
[Contents](#Contents)

```python
def top_10_longest_recipes(df):
    top_longest = df.nlargest(10, 'recipe_length')
    
    plt.figure(figsize=(10, 6))
    sns.barplot(x='recipe_length', y='name', data=top_longest, hue='name', palette='deep', dodge=False, legend=False)
    plt.xlabel('Recipe Length (Number of Words)')
    plt.ylabel('Recipe Name')
    plt.title('Top 10 Longest Recipes')
    plt.grid(axis='x', linestyle='--', linewidth=0.7)
    
    # Adding the labels on top of the bars
    for index, value in enumerate(top_longest['recipe_length']):
        plt.text(value - (value * 0.1), index, f'{value:.0f}', va='center', ha='center', color='black', weight='bold')
    
    plt.show()

# Call the function
top_10_longest_recipes(df)
```

**Chart Highlights:**
The chart titled "Top 10 Longest Recipes" showcases the recipes with the highest word counts in their instructions:
* All the top recipes are desserts, indicating that dessert recipes tend to be more detailed and elaborate in their preparation steps. This can be due to the precision required in baking and dessert-making, which often involves multiple steps and detailed instructions.
* The longest recipe, "Traditional Layered French Croissants," has 723 words, while the 10th longest, "Coconut-Rum Cake with Lemon Curd and White Chocolate Ganache," has 557 words.

This trend highlights the complexity and detailed nature of dessert recipes compared to other categories.

### **Is there a relationship between recipe length and preparation time?**<a id='Is_there_a_relationship_between_recipe_length_and_preparation_time'></a>
[Contents](#Contents)

```python
import re
def parse_time(time_str):
    """
    Parse a time string like '1 hr 25 mins' into total minutes.
    """
    if pd.isnull(time_str):
        return 0
    
    time_str = time_str.strip().lower()
    time_str = re.sub(r'[^0-9hrsmins ]', '', time_str)
    time_str = time_str.replace('s ', 's').replace(' min', ' mins').replace(' hr', ' hrs')
    time_str = time_str.replace('minss', 'mins').replace('hrss', 'hrs')

    hours = 0
    minutes = 0
    
    try:
        if 'hrs' in time_str:
            parts = time_str.split('hrs')
            hours_part = parts[0].strip()
            hours = int(hours_part) if hours_part else 0
            if len(parts) > 1 and 'mins' in parts[1]:
                minutes_part = parts[1].replace('mins', '').strip()
                minutes = int(minutes_part) if minutes_part else 0
        elif 'mins' in time_str:
            minutes = int(time_str.replace('mins', '').strip())
        return hours * 60 + minutes
    except ValueError as e:
        print(f"Error parsing time: {time_str} -> {e}")
        return 0

# Apply the time parsing function
df['prep_time_mins'] = df['prep'].apply(parse_time)

# Function to plot the relationship and show the correlation
def relation_length_prep_time(df):
    correlation = df[['recipe_length', 'prep_time_mins']].corr().iloc[0, 1]
    print(f'Correlation between Recipe Length and Preparation Time: {correlation:.2f}')

    plt.figure(figsize=(10, 6))
    sns.scatterplot(x='recipe_length', y='prep_time_mins', data=df, s=50, alpha=0.6)
    
    # Adding the correlation line
    sns.regplot(x='recipe_length', y='prep_time_mins', data=df, scatter=False, color='red', line_kws={'linewidth': 1.5})
    
    plt.xlabel('Recipe Length (Number of Words)')
    plt.ylabel('Preparation Time (Minutes)')
    plt.title(f'Recipe Length vs Preparation Time\nCorrelation: {correlation:.2f}')
    plt.grid(axis='both', linestyle='--', linewidth=0.7)
    plt.show()

# Call the function
relation_length_prep_time(df)
```

**Chart Highlights:**
* Moderate Positive Correlation: The correlation between recipe length and preparation time is 0.33, indicating a moderate positive relationship.
* Dense Cluster at Low Values: Most recipes have shorter preparation times and recipe lengths, with a dense cluster below 200 words and 100 minutes.
* Outliers: There are some outliers with very high preparation times, exceeding 600 minutes, and some recipes with lengths over 500 words.
* Trend Visibility: While the overall trend shows a positive correlation, the scatter plot reveals significant variability, especially for recipes with longer lengths.

This visualization helps to understand how preparation time generally increases with recipe length, but the relationship is not very strong, indicating other factors might also influence preparation time.

### **What is the average prep/cook/total time?**<a id='What_is_the_average_prep_cook_total_time'></a>

```python
# Apply the time parsing function
df['total_time_mins'] = df['prep_time_mins'] + df['cook_time_mins']

def average_times(df):
    avg_prep_time = df['prep_time_mins'].mean()
    avg_cook_time = df['cook_time_mins'].mean()
    avg_total_time = df['total_time_mins'].mean()

    times = {
        'Average Prep Time': avg_prep_time,
        'Average Cook Time': avg_cook_time,
        'Average Total Time': avg_total_time
    }

    plt.figure(figsize=(10, 6))
    sns.barplot(x=list(times.values()), y=list(times.keys()), hue=list(times.keys()), dodge=False, palette='deep')
    plt.xlabel('Average Time (Minutes)')
    plt.ylabel('Time Type')
    plt.title('Average Preparation, Cooking, and Total Time')
    plt.grid(axis='x', linestyle='--', linewidth=0.7)
    
    # Adding the labels on top of the bars
    for index, value in enumerate(times.values()):
        plt.text(value - (value * 0.1), index, f'{value:.2f}', va='center', ha='center', color='black', weight='bold')
    
    plt.show()

# Call the function
average_times(df)
```

### **What are the top 10 most popular ingredients?**<a id='What_are_the_top_10_most_popular_ingredients'></a>
[Contents](#Contents)

```python
from collections import Counter
import itertools


def top_10_popular_ingredients(df):
    # Combine all ingredients lists into a single list
    all_ingredients = list(itertools.chain.from_iterable(df['high_level_ingredients'].dropna().apply(eval)))
    
    # Count the occurrences of each ingredient
    ingredient_counts = Counter(all_ingredients)
    
    # Get the top 10 most common ingredients
    most_common_ingredients = ingredient_counts.most_common(10)
    
    # Separate the ingredient names and their counts for plotting
    ingredients, counts = zip(*most_common_ingredients)
    
    # Plotting
    plt.figure(figsize=(10, 6))
    sns.barplot(x=list(counts), y=list(ingredients), hue=list(ingredients), dodge=False, palette='deep', legend=False)
    plt.xlabel('Count')
    plt.ylabel('Ingredient')
    plt.title('Top 10 Most Popular Ingredients')
    plt.grid(axis='x', linestyle='--', linewidth=0.7)
    
    # Adding the labels on top of the bars
    for index, value in enumerate(counts):
        plt.text(value, index, f'{value}', va='center', ha='center', color='black', weight='bold')
    
    plt.show()

top_10_popular_ingredients(df)
```

**Chart Highlights:**
* Salt is the most popular ingredient, appearing in 13,522 recipes.
* Black pepper follows as the second most common ingredient, with 7,652 occurrences.
* Butter is third, used in 6,381 recipes, highlighting its essential role in cooking and baking.
* All-purpose flour and white sugar are also frequently used, with counts of 6,072 and 6,002, respectively, indicating their importance in various recipes.
* Onion and olive oil are notable ingredients in savory dishes, found in 5,114 and 4,968 recipes.
* Water, while commonly available, appears in 4,725 recipes, underscoring its utility in cooking processes.
* Eggs and cloves complete the top 10 list, with 3,896 and 3,720 occurrences, showing their significance in both baking and savory dishes.

These ingredients are staples in many kitchens and form the foundation of a wide variety of recipes.
